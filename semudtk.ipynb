```r
dataset_clean <- dataset[dataset$BMI != 0 & dataset$Glucose != 0 & dataset$Insulin != 0, ]
dataset_clean <- na.omit(dataset_clean)
```

```r
dataset_clean$BMI_group <- cut(dataset_clean$BMI, 
                               breaks = c(-Inf, 18.5, 24.9, 29.9, Inf), 
                               labels = c("Underweight", "Normal", "Overweight", "Obese"))

# Kiểm tra các nhóm BMI
table(dataset_clean$BMI_group)
```

```r
shapiro.test(dataset_clean$Glucose[dataset_clean$BMI_group == "Normal"]) # Nhóm Normal
shapiro.test(dataset_clean$Glucose[dataset_clean$BMI_group == "Overweight"]) # Nhóm Overweight
shapiro.test(dataset_clean$Glucose[dataset_clean$BMI_group == "Obese"]) # Nhóm Obese
```

```r
F <- table(dataset_clean$BMI_group, dataset_clean$Outcome)

# Kiểm định Chi-square giữa BMI_group và Outcome
chisq_test <- chisq.test(F)

# Xem kết quả kiểm định Chi-square
chisq_test
```

```r
shapiro.test(dataset_clean$BMI[dataset_clean$Outcome == 0])

# Kiểm tra phân phối chuẩn của BMI cho nhóm Outcome = 1 (bị bệnh)
shapiro.test(dataset_clean$BMI[dataset_clean$Outcome == 1])
```

```r
boxplot(BMI ~ Outcome, data = dataset_clean, 
        main = "So sánh BMI giữa các nhóm Outcome", 
        xlab = "Outcome (0: Không bệnh, 1: Bị bệnh)", 
        ylab = "BMI", 
        col = c("skyblue", "lightcoral"))
```

```r
mann_whitney_test <- wilcox.test(BMI ~ Outcome, data = dataset_clean)

# Xem kết quả kiểm đị
mann_whitney_test
```

p-value < 0.05,có sự khác biệt có ý nghĩa thống kê giữa hai nhóm (trong trường hợp này, giữa Outcome = 0 và Outcome = 1 về BMI)

```r
table(dataset_clean$BMI_group)
```

```r
mann_whitney_result_normal <- wilcox.test(BMI ~ Outcome, data = dataset_clean[dataset_clean$BMI_group == "Normal", ])

# Kiểm định Mann-Whitney U test cho nhóm "Overweight"
mann_whitney_result_overweight <- wilcox.test(BMI ~ Outcome, data = dataset_clean[dataset_clean$BMI_group == "Overweight", ])

# Kiểm định Mann-Whitney U test cho nhóm "Obese"
mann_whitney_result_obese <- wilcox.test(BMI ~ Outcome, data = dataset_clean[dataset_clean$BMI_group == "Obese", ])

# Xem kết quả kiểm định
mann_whitney_result_normal
mann_whitney_result_overweight
mann_whitney_result_obese
```
