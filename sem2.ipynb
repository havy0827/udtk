```r
library(tidyverse)

# Dữ liệu từ percentage_age_outcome
df <- data.frame(
  Age_group = c("Trẻ", "Trưởng thành", "Già"),
  X0 = c(93.939394, 68.253968, 36.363636),   # 0 = Không bệnh
  X1 = c(6.060606, 31.746032, 63.636364)     # 1 = Bị bệnh
)

df_long <- df %>%
  rename(`0` = X0, `1` = X1) %>%
  pivot_longer(cols = c(`0`, `1`),
               names_to = "Outcome",
               values_to = "Percent") %>%
  mutate(
    Age_num = as.numeric(factor(Age_group,
                                levels = c("Trẻ", "Trưởng thành", "Già"))),
    # Đổi 0/1 thành nhãn + đặt thứ tự: Không bệnh (dưới), Bị bệnh (trên)
    Outcome = recode(Outcome,
                     `0` = "Không bệnh",
                     `1` = "Bị bệnh"),
    Outcome = factor(Outcome, levels = c("Không bệnh", "Bị bệnh"))
  ) %>%
  arrange(Outcome, Age_num)   # sắp theo thứ tự Outcome trước

ggplot(df_long,
       aes(x = Age_num, y = Percent, fill = Outcome, group = Outcome)) +
  geom_area(alpha = 0.8, position = "stack") +
  scale_x_continuous(
    breaks = df_long$Age_num[!duplicated(df_long$Age_num)],
    labels = c("Trẻ", "Trưởng thành", "Già")
  ) +
  labs(
    title = "Nhóm 6",
    x = "Nhóm tuổi",
    y = "Tỉ lệ (%)",
    fill = "Tình trạng"
  ) +
  theme_minimal(base_size = 14)
```

```r
library(tidyverse)

# Dữ liệu từ percentage_preg_outcome
df_preg <- data.frame(
  Preg_group = c(
    "Không sinh",
    "Sinh 1 con",
    "Sinh 2–4 con",
    "Sinh 5–10 con",
    "Sinh nhiều hơn 10"
  ),
  X0 = c(66.07143, 79.61783, 65.27778, 53.84615, 25.00000),  # Không bệnh
  X1 = c(33.92857, 20.38217, 34.72222, 46.15385, 75.00000)   # Bị bệnh
)

df_preg_long <- df_preg %>%
  rename(`0` = X0, `1` = X1) %>%
  pivot_longer(cols = c(`0`, `1`),
               names_to = "Outcome",
               values_to = "Percent") %>%
  mutate(
    Preg_num = as.numeric(factor(Preg_group,
                                 levels = c("Không sinh",
                                            "Sinh 1 con",
                                            "Sinh 2–4 con",
                                            "Sinh 5–10 con",
                                            "Sinh nhiều hơn 10"))),

    # Đặt đúng nhãn + thứ tự chồng (Không bệnh dưới — Bị bệnh trên)
    Outcome = recode(Outcome,
                     `0` = "Không bệnh",
                     `1` = "Bị bệnh"),
    Outcome = factor(Outcome, levels = c("Không bệnh", "Bị bệnh"))
  ) %>%
  arrange(Outcome, Preg_num)

# Vẽ biểu đồ miền
ggplot(df_preg_long,
       aes(x = Preg_num, y = Percent, fill = Outcome, group = Outcome)) +
  geom_area(alpha = 0.8, position = "stack") +
  scale_x_continuous(
    breaks = df_preg_long$Preg_num[!duplicated(df_preg_long$Preg_num)],
    labels = c("Không sinh",
               "Sinh 1 con",
               "Sinh 2–4 con",
               "Sinh 4–10 con",
               "Sinh > 10 con")
  ) +
  labs(
    title = "Nhóm 6",
    x = "Số lần mang thai",
    y = "Tỉ lệ (%)",
    fill = "Tình trạng"
  ) +
  theme_minimal(base_size = 14)
```
