**Xét tương quan giữa BMI và T2DM**

```r
dt$BMI_group <- cut(dt$BMI, 
                               breaks = c(18.5, 24.9, 29.9, Inf), 
                               labels = c("Normal", "Overweight", "Obese"))

table(dt$BMI_group)
```

```r
F <- table(dt$BMI_group, dt$Outcome)
F
```

```r
chisq_test <- chisq.test(F)
chisq_test
```

```r
table_bmi_outcome <- table(dt$BMI_group, dt$Outcome)

# Tính % người mắc bệnh trong mỗi nhóm BMI
percentage_bmi_outcome <- prop.table(table_bmi_outcome, 1) * 100

# In kết quả
percentage_bmi_outcome
```

```r
library(ggplot2)
library(tidyr) 
library(dplyr)

df_summary <- data.frame(
  Nhom_BMI = c("Normal", "Overweight", "Obese"),
  Khong_benh = c(95.45, 78.82, 58.02), 
  Bi_benh = c(4.55, 21.18, 41.98)     
)


df_summary$Nhom_BMI <- recode(df_summary$Nhom_BMI, 
                              "Normal" = "Bình thường", 
                              "Overweight" = "Thừa cân", 
                              "Obese" = "Béo phì")

plot_data <- df_summary %>%
  pivot_longer(cols = c("Khong_benh", "Bi_benh"), 
               names_to = "Tinh_trang", 
               values_to = "Ty_le")


plot_data$Nhom_BMI <- factor(plot_data$Nhom_BMI, 
                             levels = c("Bình thường", "Thừa cân", "Béo phì"))

plot_data$Tinh_trang <- factor(plot_data$Tinh_trang, 
                               levels = c("Khong_benh", "Bi_benh"))


ggplot(plot_data, aes(x = Nhom_BMI, y = Ty_le, fill = Tinh_trang)) +
  
  geom_col(width = 0.8) + 
  geom_text(aes(label = round(Ty_le, 2)), 
            position = position_stack(vjust = 0.5), # Căn giữa vùng màu
            size = 4) +

  scale_fill_manual(values = c("Bi_benh" = "#00BFC4", "Khong_benh" = "#F8766D"),
                    labels = c("Bị bệnh", "Không bệnh")) +

  labs(title = "Nhóm 6", 
       x = "Nhóm BMI", 
       y = "Tỉ lệ (%)",
       fill = "Tình trạng") +
  

  theme_minimal() +
  theme(
    plot.title = element_text(hjust = 0, size = 14), 
    axis.text = element_text(color = "black"),
    panel.grid.major.x = element_blank() 
  )
```
